```python
# Importando as bibliotecas necessrias
from fpdf import FPDF
import pandas as pd
```

## Leitura dos Dados

Vamos ler os dados de um arquivo CSV que contm os nomes dos participantes que iro receber os certificados.

Clique no link se deseja entender sobre esta parte do cdigo: [LINK](https://)

```python
# Lendo os dados do arquivo CSV
dados = pd.read_csv("dados.csv")

# Exibindo os primeiros registros do arquivo
dados.head()
```

## Configurao dos Textos do Certificado

Definimos os textos fixos que sero inseridos nos certificados, como o ttulo, subttulo e outras informaes relevantes.

Clique no link se deseja entender sobre esta parte do cdigo: [LINK](https://)

```python
# Dados do certificado
titulo = "CERTIFICADO DE PARTICIPAO"
subtitulo = "Este certificado comprova que"
texto2 = "concluiu com xito o curso GRATUITO DE PYTHON ministrado por"
texto3 = "PROF. SAUER entre xx/xx/xxxx e xx/xx/xxxx,"
texto4 = "com carga horria de aproximadamente xx horas."
```

## Gerao dos Certificados

Para cada nome presente no arquivo CSV, vamos criar um novo certificado em PDF. Usamos a biblioteca `FPDF` para manipular o PDF, adicionar texto e imagens.

Clique no link se deseja entender sobre esta parte do cdigo: [LINK](https://)

```python
# Loop para gerar certificados para cada participante
for nome in dados['nomecompleto']:
    # Criao do objeto PDF
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", 'B', size=15)

    # Adicionando a imagem de fundo (template)
    pdf.image("template.png", x=0, y=0)

    # Configurando a cor do texto
    pdf.set_text_color(33, 24, 136)

    # Adicionando o texto ao PDF
    pdf.text(65, 95, titulo)
    pdf.text(67, 120, subtitulo)
    pdf.text(70, 145, nome)
    pdf.text(20, 165, texto2)
    pdf.text(50, 175, texto3)
    pdf.text(50, 185, texto4)

    # Salvando o PDF com um nome nico
    pdf.output(f"Certificado_{nome}.pdf")
```

## Concluso

Este notebook demonstrou como gerar certificados personalizados em PDF para um grupo de participantes usando dados de um arquivo CSV. Utilizamos a biblioteca `FPDF` para criar e manipular os PDFs, e a biblioteca `pandas` para ler e manipular os dados do CSV.

## Cdigo Completo do Projeto

```python
from fpdf import FPDF
import pandas as pd

dados = pd.read_csv("dados.csv")
#print(dados['nomecompleto'])

# Dados do certificado
titulo =    "CERTIFICADO DE PARTICIPAO"
subtitulo = "Este certificado comprova que"
nome = "Alexandre Sauer"
texto2 = "concluiu com xito o curso GRATUITO DE PYTHON ministrado por"
texto3 = " PROF. SAUER entre 23/08/2024 e 28/08/2024,"
texto4 = "com carga horria de aproximadamente 08 horas."


for nome in dados['nomecompleto']:
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", 'B', size=15)
    pdf.image("template.png", x=0, y=0)
    pdf.set_text_color(33, 24, 136)
    # Adicionar o texto
    # Note que as coordenadas para o texto podem precisar ser ajustadas
    pdf.text(65, 95, titulo)
    pdf.text(67, 120, subtitulo)

    pdf.text(70, 145, nome)
    pdf.text(20, 165, texto2)
    pdf.text(50, 175, texto3)
    pdf.text(50, 185, texto4)

    # Salvar o PDF com um nome de arquivo nico
    pdf.output(f"Certificado_{nome}.pdf")
```
